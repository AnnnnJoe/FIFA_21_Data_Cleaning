## Data Cleaning

### Transforming `Height` and `Weight` 

```python
# checking datatypes
print("Data type of Height: ", df['Height'].dtype)
print("Data type of Weight: ", df['Weight'].dtype)
```

Both are of the data type 'object'. We need to convert it to numeric data. 

**`Height`**

```python
# checking unique values of 'Height'
df["Height"].unique()
```

The values have 'cm' in them and some values are in inches.

Let us first remove 'cm' from the values.

```python
# removing 'cm'
df['Height'] = df['Height'].str.replace('cm', '')


df["Height"].unique()
```

Now let us convert feet and inches into cm.

We will use the below conversion:

1 ft = 30.48cm

1 in = 2.54cm

```python
# function for converting feet and inches to cm
def feet_inches_to_cm(height):
    if '\'' in height:
        feet, inches = height.split('\'')
        return round(int(feet) * 30.48 + int(inches.strip('"')) * 2.54)
    else:
        return int(height)

df['Height'] = df['Height'].apply(feet_inches_to_cm)
```

```python
df['Height'].unique()
```

**`Weight`**

```python
# checking unique values of 'Weight'
df.Weight.unique()
```

Some values are in pounds and some in kg. Let us make it all consistent and convert to kg.

First, let us remove 'kg' from the values

```python
# removing 'kg'
df['Weight'] = df['Weight'].str.replace('kg', '')

df["Weight"].unique()
```

Now let us convert pound to kg.

We will use the below conversion:

1 lbs = 0.45359237 kg

```python
# converting pounds to kg
def pound_to_kg(weight):
    if 'lbs' in weight:
        weight = weight.strip('lbs') # remove 'lbs' from the values
        return round(int(weight) * 0.45359237)
    else:
        return int(weight)

df['Weight'] = df['Weight'].apply(pound_to_kg)
```

```python
df['Weight'].unique()
```

Let us check the datatypes again.

```python
print("Data type of Height: ", df['Height'].dtype)
print("Data type of Weight: ", df['Weight'].dtype)
```

All the values are consistent and in the same units. 

Finally, let us change the column names for more clarity.

```python
# renaming the columns
df.rename(columns={'Height': 'Height(cm)', 'Weight':'Weight(kg)'}, inplace=True)
```

### Transforming `Value`, `Wage`, `Release Clause` 

```python
# checking datatypes of each column
print('Value: ', df['Value'].dtype)
print('Wage: ', df['Wage'].dtype)
print('Release Clause: ', df['Release Clause'].dtype)
```

These columns should be of numeric datatypes as it represents market value, weekly wage, and release clause of each player.

To do this, we will first need to remove the `€` symbol from the values.

```python
# removing '€' symbol
df['Value'] = df['Value'].str.replace('€','')
df['Wage'] = df['Wage'].str.replace('€','')
df['Release Clause'] = df['Release Clause'].str.replace('€','')
```

Now we can change the columns to numeric datatype.

```python
# function for converting string to number
def str_to_num(value):
    if 'K' in value:
        return round(float(value.strip('K')) * 1000)
    elif 'M' in value:
        return round(float(value.strip('M')) * 1000000)
    else:
        return int(value)

df['Value'] = df['Value'].apply(str_to_num)
df['Wage'] = df['Wage'].apply(str_to_num)
df['Release Clause'] = df['Release Clause'].apply(str_to_num)
```

```python
# checking datatypes of each column
print('Value: ', df['Value'].dtype)
print('Wage: ', df['Wage'].dtype)
print('Release Clause: ', df['Release Clause'].dtype)
```

### Transforming `Club`

### Creating new column `'Agreement'`

Let us now create a new column `'Agreement'` based on the column 'Contract'. 'Contract' contains different types of values representing paid agreements, loan agreements, and free agent statuses for different players. Let us group these into **'Contract'**, **'Loan'**, and **'Free'**.

```python
# function for the condition
def contract_type(contract):
    if '~' in contract:
        type = 'Contract'
    elif 'Loan' in contract:
        type = 'Loan'
    else:
        type = 'Free'
    return type
        
df['Agreement'] = df['Contract'].apply(contract_type)
```

```python
df.Agreement.unique()
```

### Transforming `W/F`, `SM`, `IR`

**W/F ("Weak Foot")**: It refers to the rating or assessment of a player's weaker foot, indicating their proficiency in using their non-dominant foot for activities such as passing, shooting, and dribbling. The rating is often represented as a number between 1 and 5, with 5 being the highest proficiency.

**SM ("Skill Moves")**: "Skill Moves" refer to the range of advanced techniques and tricks that a player can perform in the game to outmaneuver opponents. The "Skill Moves" rating usually ranges from 1 to 5 or 1 to 5 stars, indicating the complexity and effectiveness of the player's skill moves repertoire.

**IR (International Reputation)**: "International Reputation" refers to the perceived prestige and recognition of a player on the international stage. It typically reflects the player's reputation based on their performance in international competitions and their standing among football fans worldwide. The "International Reputation" rating ranges from 1 to 5 or 1 to 5 stars, with higher values indicating greater recognition and prestige.

`W/F`, `SM`, and `IR` ranges between 1 - 5. These values also have a "★" character in it. Let us remove this character from the columns. 

Let us also change the column names for improved clarity and readability.

```python
# removing '★'
df['W/F'] = df['W/F'].str.replace('★', '')
df['SM'] = df['SM'].str.replace('★', '')
df['IR'] = df['IR'].str.replace('★', '')
```

```python
# removing white spaces
df['W/F'] = df['W/F'].str.strip()
df['SM'] = df['SM'].str.strip()
df['IR'] = df['IR'].str.strip()
```

```python
# changing datatype to numeric
df['W/F'] = pd.to_numeric(df['W/F'])
df['SM'] = pd.to_numeric(df['SM'])
df['IR'] = pd.to_numeric(df['IR'])
```

```python
# changing column names
df.rename(columns={'W/F': 'Weak Foot', 'SM': 'Skill Moves', 'IR': 'Intl Rep'}, inplace=True)
```

```python
# checking values & datatypes
print('Weak Foot')
print(df['Weak Foot'].unique())
print(df['Weak Foot'].dtype)

print('----------------------')

print('Skill Moves')
print(df['Skill Moves'].unique())
print(df['Skill Moves'].dtype)

print('----------------------')

print('Intl Rep')
print(df['Intl Rep'].unique())
print(df['Intl Rep'].dtype)

print('----------------------')
```

### Transforming `Hits`

**Hits** may refer to successful passes or tackles made by players during a match. In this case, “hits” might be used to indicate successful actions.

```python
df.Hits.unique()
```

```python
# checking missing values
df.Hits.isnull().sum()
```

The `Hits` column is currently not numeric. It contains 2595 missing values. Some of the values contain suffix 'K' in them which needs to be changed.

```python
# function to convert value to number
def to_num(value):
    if pd.isnull(value): # check if value is NaN and return 0 if it is
        return 0
    elif isinstance(value, float):  # Check if value is already a float
        return int(value)  # Convert float to int
    elif 'K' in value:
        return round(float(value.strip('K')) * 1000) # removing 'K' suffix and multiplying the value with 1000
    else:
        return int(value)

# applying the function
df['Hits'] = df['Hits'].apply(to_num)
```

```python
df['Hits'].unique()
```
